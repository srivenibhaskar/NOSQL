DATA CLEANING PROCESS USING PANDAS

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("netflix_data.csv", header=0)
```

```python
df.nunique()  #analysing the number of unique values in every columns.
```

```python
plt.figure(figsize = (19,7) )
sns.heatmap(df.isna(), cmap="YlGnBu"
            ,cbar_kws={'label': 'Missing Data'}
           )
```

Through this we can see that "director" , "cast", "country" and "ratings" have missing values.

```python
df.head()
```

```python
# replacing the NaN values with UNKNOWN
df[ ["director", "cast"] ] = df[ ["director", "cast"] ].replace(np.nan, "UNKNOWN")
```

```python
# Find the most frequent country
most_frequent_country = df['country'].mode()[0]

# Replace NaN values in the 'country' column with the most frequent country
df['country'].fillna(most_frequent_country, inplace=True)


#same with rating.
most_frequent_rating = df['rating'].mode()[0]
df['rating'].fillna(most_frequent_rating, inplace=True)

#same with "date_Added"
most_frequent_date_added = df['date_added'].mode()[0]

# Replace NaN values in the 'country' column with the most frequent country
df['date_added'].fillna(most_frequent_date_added, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Group by the values of the rating column to check the count of the values in each column of the each rating 
rating_counts = df['rating'].value_counts().reset_index()
rating_counts.columns = ['rating', 'count']
print(rating_counts)
```

```python
# Define a mapping of ratings to categories
rating_mapping = {
    'G': 'General Audience',
    'NC-17': 'Adults Only',
    'NR': 'Not Rated',
    'PG': 'Parental Guidance',
    'PG-13': 'Parents Strongly Cautioned',
    'R': 'Restricted',
    'TV-14': 'Parents Strongly Cautioned (TV)',
    'TV-G': 'General Audience (TV)',
    'TV-MA': 'Mature Audience (TV)',
    'TV-PG': 'Parental Guidance Suggested (TV)',
    'TV-Y': 'All Children (TV)',
    'TV-Y7': 'Directed to Older Children (TV)',
    'TV-Y7-FV': 'Directed to Older Children - Fantasy Violence (TV)',
    'UR': 'Unrated'
}

df['rating_ABBR'] = df['rating'].map(rating_mapping)


```

```python
df.to_csv('netflix_data_updated.csv', index=False)
```
